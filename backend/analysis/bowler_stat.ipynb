## Group by bowler name

```python
bowlers_group = df.groupby("bowler")
```

```python
ts_mills_group = bowlers_group.get_group("TS Mills")
```

```python
ts_mills_bowled_against = ts_mills_group["batting_team"].unique().tolist()
```

```python
ts_mills_bowled_against
```

```python
def get_ts_mills_bowling_stats():
    g = ts_mills_group.groupby(["batting_team", "match_id", "over"])
    for team in ts_mills_bowled_against:
        print(g.get_group(team))


get_ts_mills_bowling_stats()
```
